```python
def read_file(filename):
    with open(filename) as file:
        data = file.read()
        return data
```

```python
codon_raw_values = read_file("RNACodonTable.txt")
```

```python
def parse_contents(contents):
    # Split contents into non whitespace values
    contents = contents.split()
    # The dictionary that will hold all of our key/value pairs
    result_dict = {}
    # When stepping through the list by 2 starting from the first index we collect all of the keys
    keys = contents[::2]
    # When stepping through the list by 2 starting from the second index we collect all of the values
    vals = contents[1::2]
    # For each key and value pair assign our dictionary
    # For this to work well the length of keys must be equal to the length of values
    for index in range(len(keys)):
        result_dict[keys[index]] = vals[index]
    return result_dict
```

```python
codon_dictionary = parse_contents(codon_raw_values)
```

```python
def get_substring_slices(given, slice_length=3):
    resultant = []
    for i in range(0,len(given),slice_length):
        resultant.append(given[i:i+slice_length])
    return resultant
```

```python
def answer(RNAString, codon_dict):
    slices = get_substring_slices(RNAString,3)
    resultant_str = ''
    for elem in slices:
        decoded_mrna = codon_dict[elem]
        if decoded_mrna == 'Stop':
            break
        resultant_str += decoded_mrna
    return resultant_str
```
